Agglomerative clustering is a type of hierarchical clustering that builds the clusters in a bottom-up manner.Initially,each data point is treated as a seperate cluster.Then,in each step the two closest clusters are merged based on a chosen distance metric.This process continues until the desired number of clusters is formed or all points are merged into one big cluster.

```python
#importing
from sklearn.cluster import AgglomerativeClustering
from sklearn import datasets
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

#giving datasets
data,_= make_blobs(n_samples=300,centers=4,cluster_std=0.60,random_state=0)

#clustering
agg_cluster=AgglomerativeClustering(n_clusters=4,metric="euclidean",linkage="ward")
clusters=agg_cluster.fit_predict(data)

#plotting
plt.scatter(data[:,0],data[:,1],c=clusters,cmap="jet",edgecolor='k')
plt.title("Agglomerative Clustering",fontsize=16)
plt.ylabel("Feature 2")
plt.show()
```
